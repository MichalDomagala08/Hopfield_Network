```python
### == Utilities ===

import numpy as np

#Zamienia reprezentacje 0,1 na -1,1
def dychotomy(wektor):
    op = np.zeros(wektor.shape[0])

    for i in range(wektor.shape[0]):
        if wektor[i] ==0:
            op[i] = -1
        else:
            op[i] = 1
    
    return op

#decode - zamienia reprezentację -1,1 na 0,1
def decode(rray):
    eq = np.zeros((rray.shape[0]),dtype='int')
    #print(eq)
    for i in range(len(rray)):
        if rray[i] == 1:
            eq[i] = 1
    return np.matrix(eq,dtype='int')

# Ogarnięcie znaku zależnie od znaku sumy
def signum(sums):
    sumss = np.array(sums)


    sums_eq = np.zeros(sums.shape[1])

    for i in range(len(sumss[0])):
        #print(sums_eq)

        if sumss[0][i] >= 0:
            sums_eq[i] = 1
        else:
            sums_eq[i] = -1
        
    return sums_eq



#Funkcja Ograniczenia pamięci
def memory_limit_function(data):
    '''
    
        Funkcja Ograniczająca Pamięć. innymi słowy, 
        h dane przechowywane w pamięci nie mogą być większe 
        od podłogi z wymiarowości każego wektora danych podzielonego przez podwojony logarytma z wymiarowości
    '''
    import math
    
    return data.shape[0] <= math.floor(data.shape[1]/(2*math.log(data.shape[1])))
    
#print( memory_limit_function(data))

#Funkcja obliczająca Energię sieci
def energy(weights,x):
    
    """
        Energia to malejąca funckaj która opisuje komputacje w sieci.Jeśli ta funkcja
osiągnie minimum lokalne, to sieć Hopfielda zakończy odzyskiwanie wektora z macierzy

        mnożymy po prostu wektor rtransponowany razy macierz Wag razy weektor i razy -1/2
    """
    E = -(1/2)*x.T.dot(weights).dot(x)
    return np.array(E)
```

```python
import numpy as np
import random

#Funckja Wagowa
def weights(wektory):
    """
        Wagi ustala się nożac wektor początkowy przez samego siebie,
        zamienajć 0 na Diagonalu Macierzy
    """
    #print( wektory.T.dot( wektory))
    xx = wektory.T.dot( wektory)
    np.fill_diagonal(xx,0)
    return xx


    

#input_wektor = np.array([-1,1,1,-1,1])
def recovery_synchronus(weights,inp):
    """
        Odzyskiwanie Synchroniczne Danych z macierzy
        
        Zwyczjanie mnożymy wektor wejsciowy przez Macirz Wag, aktywując każdy z nuronów
        
        następnie ustalamy znak, który zwracamy jako nasz Wektor
    """
    sums  =np.matmul(inp,weights)
    sums_eq = signum(sums)
        
    return sums_eq



def recovery_asynchronus(weights,inp, it = 10000):
    """
    
    Odzyskiwanie Asynchronicze:
    
    Iteracyjnie odzyskujemy dane z Sieci, 
    przez aktywacje psozczególnego nerunu prze wektor v'
    
    1) wybieramy wektor losowo, 
    2) mnożymy wektor przez nasz wektor stanu, 
    3) ustalamy stan neuronu w wektorze stanu 
       jako wynik signum z wyniku mnożenia
    
    """
    weights = np.array(weights)
    inp = np.array(inp)
    prev_en = 0
    En = 0
    count =0
    wh = True
    print('Outputs: ')
    while wh == True and count < 10000:
        count +=1
        column = random.randint(0,inp.shape[0] -1)
        Value = np.sign(weights[:][column].dot(inp.T))
        
        ###Checking####
        #print('Picked vector: {}'.format(column))
        #print('Calculated value: {}'.format(Value))
        #print('Updated vector:\n {}'.format(inp))'''
        inp[column] = Value
        prev_en = En
        En = energy(weights,inp)
        
        ####Checking 2 ####
        #print('Energy: {}'.format(En))
        #print()
        #middle = decode(inp)
        periodic_draw(inp,En) 
        if prev_en <= -400 :
            
            ####Kolejne PEriodyczne Sprawdzenie###
            #print('Energy: {}'.format(En))
            #print('Count: {}'.format(count))
            
            return inp
            wh == False
            break
            pass
        
    ###Periodyczne sprawdzanie Energii
    
    print(inp)
    return inp
```
